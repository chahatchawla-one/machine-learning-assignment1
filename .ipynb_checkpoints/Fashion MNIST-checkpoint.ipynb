# Load the libraries

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from mnist import MNIST

from Modules.Distributions import Gaussian1D
from Modules.Metrics import Accuracy, Precision, Recall, F1score
from Modules.SupervisedLearning import NaiveBayesClassifier, BayesClassifier, kNNClassifier, ParzenWindow
from Modules.UnsupervisedLearning import kmeansClassifier
from Modules.Visualizers import VisualizePCA, PCA
```

# Load Fashion MNIST dataset

```python
fashion_data = MNIST("data/fashion_mnist")

x_train, y_train = fashion_data.load_training()
x_train = np.array(x_train)
y_train = np.array(y_train)

x_test, y_test = fashion_data.load_testing()
x_test = np.array(x_test)
y_test = np.array(y_test)

# plt.imshow(x_train[0].reshape(28, 28), cmap = "Greys_r")

# p = 50 for Naive Bayes
# p = 70 for Bayes
```

```python
VisualizePCA(x_train)
```

```python
x_train, x_test, p = PCA(x_train, x_test, p = 50)
```

# Naive Bayes classifier (MLE)

```python
num_alphas = {
    0 : np.ones(p).astype(int),
    1 : np.ones(p).astype(int),
    2 : np.ones(p).astype(int),
    3 : np.ones(p).astype(int),
    4 : np.ones(p).astype(int),
    5 : np.ones(p).astype(int),
    6 : np.ones(p).astype(int),
    7 : np.ones(p).astype(int),
    8 : np.ones(p).astype(int),
    9 : np.ones(p).astype(int)
}

classifier = NaiveBayesClassifier(algorithm = "MLE")
classifier.Fit(x_train, y_train, num_alphas, roc_plot = True)
print("Training accuracy:", classifier.training_accuracy)

predictions = classifier.Predict(x_test)
print("Test Accuracy =", Accuracy(predictions, y_test))
print("Precision =", Precision(predictions, y_test))
print("Recall =", Recall(predictions, y_test))
print("F1score =", F1score(predictions, y_test))
```

## Bayes classifier (MLE)

```python
mu_ = x_train.mean(axis = 0)
sigma_ = np.sqrt(x_train.var(axis = 0))
x_train_ = (x_train - mu_) / sigma_
x_test_ = (x_test - mu_) / sigma_

num_alphas = {
    0 : 1,
    1 : 1,
    2 : 1,
    3 : 1,
    4 : 1,
    5 : 1,
    6 : 1,
    7 : 1,
    8 : 1,
    9 : 1
}

classifier = BayesClassifier(algorithm = "MLE")
classifier.Fit(x_train_, y_train, num_alphas)
print("Training accuracy:", classifier.training_accuracy)

predictions = classifier.Predict(x_test_)
print("Test Accuracy =", Accuracy(predictions, y_test))
print("Precision =", Precision(predictions, y_test))
print("Recall =", Recall(predictions, y_test))
print("F1score =", F1score(predictions, y_test))
```

# Parzen window (hypercube window)

```python
# p = 70

h_array = np.linspace(120, 1000, num = 45)
accuracy = pd.DataFrame()
accuracy["h"] = h_array
acc = []
for h in h_array:
    classifier = ParzenWindow(h = h)
    classifier.Fit(x_train, y_train)
    acc_ = Accuracy(classifier.Predict(x_test), y_test)
    acc.append(acc_)
    print(h)
    print(acc_)
accuracy["accuracy"] = acc
```

```python
accuracy
```

```python
accuracy
```

```python
fig, ax = plt.subplots()
accuracy.plot(x = "h", y = "accuracy", figsize = [25, 15], ax = ax)
plt.xticks(h_array)
ax.grid(linestyle = "-", linewidth = '0.2', color='black')
plt.title("Fashion MNIST data Parzen window (hypercube) performance vs window width")
```

# Parzen window (Gaussian window)

```python
# p = 70

h_array = np.linspace(120, 1000, num = 45)
accuracy = pd.DataFrame()
accuracy["h"] = h_array
acc = []
for h in h_array:
    classifier = ParzenWindow(h = h, kernel = "Gaussian")
    classifier.Fit(x_train, y_train)
    acc.append(Accuracy(classifier.Predict(x_test), y_test))
    print(h)
accuracy["accuracy"] = acc
```

```python
accuracy
```

```python
fig, ax = plt.subplots()
accuracy.plot(x = "h", y = "accuracy", figsize = [25, 15], ax = ax)
plt.xticks(h_array)
ax.grid(linestyle = "-", linewidth = '0.2', color='black')
plt.title("Fashion MNIST data Parzen window (Gaussian) performance vs window width")
```

# kNN

```python
# p = 75

distance_metrics = ["Euclidean", "Manhattan", "Chebyshev", "Minkowski"]

accuracy = pd.DataFrame()
accuracy["k"] = range(1, 11)

precision = pd.DataFrame()
precision["k"] = range(1, 11)

recall = pd.DataFrame()
recall["k"] = range(1, 11)

f1_score = pd.DataFrame()
f1_score["k"] = range(1, 11)

for j in distance_metrics:
    classifier = kNNClassifier(distance_metric = j)

    accuracy_ = []
    precision_ = []
    recall_ = []
    f1_score_ = []
    for k in range(1, 11):
        predictions = classifier.Predict(x_train, y_train, x_test, k)
        predictions = np.array(predictions)
        accuracy_.append(Accuracy(predictions, y_test))
        precision_.append(Precision(predictions, y_test))
        recall_.append(Recall(predictions, y_test))
        f1_score_.append(F1score(predictions, y_test))
        
        print(k)
    
    accuracy[j] = accuracy_
    precision[j] = precision_
    recall[j] = recall_
    f1_score[j] = f1_score_

# from sklearn.neighbors import KNeighborsClassifier

# distance_metrics = ["euclidean", "manhattan", "chebyshev"]
# accuracy = pd.DataFrame()
# accuracy["k"] = range(1, 11)

# for j in distance_metrics:
#     acc = []
#     for k in range(1, 11):
#         classifier = KNeighborsClassifier(n_neighbors = k, metric = j)
#         classifier.fit(x_train, y_train)
#         l = classifier.predict(x_test)
#         acc.append(Accuracy(np.array(l), y_test))
#         print(h)
    
#     accuracy[j] = acc
```
